# Data Cleaning and Feature Engineering

```python
df_amazon
```

```python
df_amazon = df_amazon.drop(['date','rating'],axis=1)
```

```python
df_amazon
```

```python
variation_dummies = pd.get_dummies(df_amazon['variation'],drop_first=True)
```

```python
variation_dummies
```

```python
df_amazon.drop(['variation'],axis=1,inplace=True)
```

```python
df_amazon = pd.concat([df_amazon,variation_dummies],axis=1)
```

```python
df_amazon
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
amazon_countvectorizer = vectorizer.fit_transform(df_amazon['verified_reviews'])
```

```python
amazon_countvectorizer.shape
```

```python
print(vectorizer.get_feature_names())
```

```python
print(amazon_countvectorizer.toarray())
```

```python
df_amazon.drop(['verified_reviews'],axis=1,inplace=True)
```

```python
df_amazon
```

```python
encoded_reviews = pd.DataFrame(amazon_countvectorizer.toarray())
```

```python
df_amazon = pd.concat([df_amazon,encoded_reviews],axis=1)
```

```python
df_amazon
```

```python
X = df_amazon.drop(['feedback'],axis=1)
```

```python
y = df_amazon['feedback']
```

```python
y
```

# Model Training

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest_classifier = RandomForestClassifier(n_estimators=100,criterion='entropy')
```

```python
randomforest_classifier.fit(X_train,y_train)
```

# Model Evaluation

```python
y_predict = randomforest_classifier.predict(X_train)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_predict,y_train)
```

```python
sns.heatmap(cm,annot=True,fmt='d')
```

```python
print(classification_report(y_train,y_predict))
```

```python
y_predict_test = randomforest_classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_predict_test,y_test)
```

```python
sns.heatmap(cm,annot=True,fmt='d')
```

```python
print(classification_report(y_test,y_predict_test))
```
